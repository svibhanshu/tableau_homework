# Tableau Homework: New York City Bike Data (January and July 2018)
This homework looks at two datasets - January 2018 and July 2018 - in order to compare various aspects of the bike usage during the winter and summer months in New York. The tableau visualizations are in the form of three dashboards.

The first dashboard highlights
1. total usage of bikes during the period under study;
2. comparative usage of bikes in the months of January and July 2018;
   The bike usage during the summer month of July is much greater than the bike usage during the winter month of 
   January.
3. gender-based usage of the bikes;
   the percentage of Males using the city bikes is much greater than the percentage of females using the bikes.
4. Peak hours of the bike usage in January and July 2018; and
5. bike-usage data based on user types in the months of January and July 2018.
    As one can see that the number of customer-based usage of data is minimal compared to the subscriber-based usage, who would essentially be the local residents.

The second dashboard provides a geo-mapping of
6. top 20 starting stations of the bike usage;
7. top 20 destinations of the bike usage;

The third dashboard analyses

8. Age-based and usertype-based usage of the bikes, 
   The age-based data is divided into three groups - (a)Below 30, (b)Age (30-45) and above 45
9. Bottom 20 starting stations of the bike usage
10. Bottom 20 destinations of the bike usage


```python
import pandas as pd
```

```python
# Loading the January 2018 Data
df_jan2018=pd.read_csv("Jan2018.csv")
df_jan2018.head()
```

```python
# Loading the July 2018 Data
df_july2018=pd.read_csv("July2018.csv")
df_july2018.head()
```

```python
# Combine the two datasets using the concatinating function
df_2018=pd.concat([df_jan2018, df_july2018])
df_2018.head()
```

```python
# Saving a csv file of the combined dataset
df_2018.to_csv("bike_data.csv",index=False)
```
